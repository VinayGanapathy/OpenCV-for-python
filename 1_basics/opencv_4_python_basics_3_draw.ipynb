```python
import cv2 as cv
import numpy as np
from PIL import Image
from google.colab.patches import cv2_imshow
```

```python
blank = np.zeros((500,500,3), dtype="uint8")
cv2_imshow(blank)
```

```python
#1. Paint the image by a certain (green) colour
blank[:] = 0,255,0
cv2_imshow(blank)
```

```python
#2. Paint the image by a certain (red) colour
blank[:] = 0,0,255
cv2_imshow(blank)
```

```python
#2. Paint the image by a certain (red) colour in a range
blank = np.zeros((500,500,3), dtype="uint8")
blank[200:300, 300:400] = 0,0,255
cv2_imshow(blank)
```

```python
# Drawing a rectangle
blank = np.zeros((500,500,3), dtype="uint8")
cv.rectangle(blank, (0,0), (250,250), (0,255,0), thickness=2)
cv2_imshow(blank)
```

```python
# Drawing a rectangle
blank = np.zeros((500,500,3), dtype="uint8")
cv.rectangle(blank, (0,0), (250,500), (0,255,0), thickness=cv.FILLED)
cv2_imshow(blank)
```

```python
# Drawing a rectangle
blank = np.zeros((500,500,3), dtype="uint8")
cv.rectangle(blank, (0,0), (250,500), (0,255,0), thickness=-1)
cv2_imshow(blank)
```

```python
# Drawing a rectangle
blank = np.zeros((500,500,3), dtype="uint8")
cv.rectangle(blank, (0,0), (blank.shape[1]//2, blank.shape[0]//2), (0,255,0), thickness=cv.FILLED)
cv2_imshow(blank)
```

```python
# Drawing a standalone line
```

```python
# Writing text on an image
```

```python
blank = np.zeros((500,500,3), dtype="uint8")
cv.putText(blank,"Hello my name is Jason!!!", (225,225), cv.FONT_HERSHEY_TRIPLEX, 1.0, (0,255,0), 2)
cv2_imshow(blank)
```

```python
blank = np.zeros((500,500,3), dtype="uint8")
cv.putText(blank,"Hello my name is Jason!!!", (0,225), cv.FONT_HERSHEY_TRIPLEX, 1.0, (0,255,0), 2)
cv2_imshow(blank)
```
