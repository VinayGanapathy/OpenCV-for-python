```python
import cv2 as cv
import numpy as np
from PIL import Image
from google.colab.patches import cv2_imshow
```

```python
park = cv.imread("/content/drive/MyDrive/Colab Notebooks/opencv_for_python/Resources/Photos/park.jpg")
cv2_imshow(park)
cv.waitKey(0) 
cv.destroyAllWindows()
```

```python
b,g,r = cv.split(park)
```

```python
cv2_imshow(b)
cv2_imshow(g)
cv2_imshow(r)
```

```python
print(park.shape)
print(b.shape)
print(g.shape)
print(r.shape)
```

```python
# Merging the color channels
merge = cv.merge([b,g,r])
cv2_imshow(merge)
```

```python
blank = np.zeros(park.shape[:2], dtype ='uint8')
```

```python
blue = cv.merge([b, blank, blank])
cv2_imshow(blue)
```

```python
green = cv.merge([blank, g, blank])
cv2_imshow(green)
```

```python
red = cv.merge([blank, blank, r])
cv2_imshow(red)
```
