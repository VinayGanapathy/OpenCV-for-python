```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
from google.colab.patches import cv2_imshow
```

```python
cats = cv.imread("/content/drive/MyDrive/Colab Notebooks/opencv_for_python/Resources/Photos/cats.jpg")
cv2_imshow(cats)
cv.waitKey(0) 
cv.destroyAllWindows()
```

```python
gray = cv.cvtColor(cats, cv.COLOR_BGR2GRAY)
cv2_imshow(gray)
```

```python
# Laplacian 
```

```python
lap = cv.Laplacian(gray, cv.CV_64F)
lap = np.uint8(np.absolute(lap))
cv2_imshow(lap)
```

```python
park = cv.imread("/content/drive/MyDrive/Colab Notebooks/opencv_for_python/Resources/Photos/park.jpg")
cv2_imshow(park)
cv.waitKey(0) 
cv.destroyAllWindows()
```

```python
gray = cv.cvtColor(park, cv.COLOR_BGR2GRAY)
cv2_imshow(gray)
```

```python
lap = cv.Laplacian(gray, cv.CV_64F)
lap = np.uint8(np.absolute(lap))
cv2_imshow(lap)
```

```python
# Sobel Gradients
```

```python
sobelx = cv.Sobel(gray, cv.CV_64F, 1, 0)
cv2_imshow(sobelx)
```

```python
sobely = cv.Sobel(gray, cv.CV_64F, 0, 1)
cv2_imshow(sobely)
```

```python
#combined_sobel1 = cv.Sobel(gray, cv.CV_64F, 1, 1)
combined_sobel2 = cv.bitwise_or(sobelx,sobely)
#cv2_imshow(combined_sobel1)
cv2_imshow(combined_sobel2)
```

```python
canny = cv.Canny (gray, 150,175)
cv2_imshow(canny)
cv2_imshow(lap)
cv2_imshow(combined_sobel2)
```
