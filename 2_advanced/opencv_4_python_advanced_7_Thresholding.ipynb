```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
from google.colab.patches import cv2_imshow
```

```python
cats = cv.imread("/content/drive/MyDrive/Colab Notebooks/opencv_for_python/Resources/Photos/cats.jpg")
cv2_imshow(cats)
cv.waitKey(0) 
cv.destroyAllWindows()
```

```python
gray = cv.cvtColor(cats, cv.COLOR_BGR2GRAY)
cv2_imshow(gray)
```

```python
# Simple Threshaolding
```

```python
threshold, thresh = cv.threshold(gray, 150, 255, cv.THRESH_BINARY)
cv2_imshow(thresh)
```

```python
threshold
```

```python
threshold, thresh = cv.threshold(gray, 100, 255, cv.THRESH_BINARY)
cv2_imshow(thresh)
```

```python
threshold, thresh = cv.threshold(gray, 225, 255, cv.THRESH_BINARY)
cv2_imshow(thresh)
```

```python
threshold, thresh_inv = cv.threshold(gray, 150, 255, cv.THRESH_BINARY_INV)
cv2_imshow(thresh_inv)
```

```python
# Adaptive Thresholding
```

```python
adaptive_thresh = cv.adaptiveThreshold(gray, 255, cv.ADAPTIVE_THRESH_MEAN_C, cv.THRESH_BINARY, 11, 3)
cv2_imshow(adaptive_thresh)
```

```python
adaptive_thresh_inv = cv.adaptiveThreshold(gray, 255, cv.ADAPTIVE_THRESH_MEAN_C, cv.THRESH_BINARY_INV, 11, 3)
cv2_imshow(adaptive_thresh_inv)
```

```python
adaptive_thresh = cv.adaptiveThreshold(gray, 255, cv.ADAPTIVE_THRESH_GAUSSIAN_C, cv.THRESH_BINARY_INV, 11, 9)
cv2_imshow(adaptive_thresh)
```
