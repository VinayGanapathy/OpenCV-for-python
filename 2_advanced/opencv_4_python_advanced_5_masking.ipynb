```python
import cv2 as cv
import numpy as np
from PIL import Image
from google.colab.patches import cv2_imshow
```

```python
cats = cv.imread("/content/drive/MyDrive/Colab Notebooks/opencv_for_python/Resources/Photos/cats.jpg")
cv2_imshow(cats)
cv.waitKey(0) 
cv.destroyAllWindows()
```

```python
blank = np.zeros(cats.shape[:2], dtype='uint8')
cv2_imshow(blank)
```

```python
mask = cv.circle(blank, (cats.shape[1]//2, cats.shape[0]//2), 100, 255, -1)
cv2_imshow(mask)
```

```python
masked_cats = cv.bitwise_and(cats, cats, mask=mask)
cv2_imshow(masked_cats)
```

```python
cats = cv.imread("/content/drive/MyDrive/Colab Notebooks/opencv_for_python/Resources/Photos/cats 2.jpg")
cv2_imshow(cats)
cv.waitKey(0) 
cv.destroyAllWindows()
```

```python
blank = np.zeros(cats.shape[:2], dtype='uint8')
mask = cv.circle(blank, (cats.shape[1]//2 + 45, cats.shape[0]//2 ), 100, 255, -1)
cv2_imshow(mask)
```

```python
masked_cats = cv.bitwise_and(cats, cats, mask=mask)
cv2_imshow(masked_cats)
```

```python
blank = np.zeros(cats.shape[:2], dtype='uint8')
circle = cv.circle(blank.copy(), (cats.shape[1]//2 + 45, cats.shape[0]//2 ), 100, 255, -1)
rectangle = cv.rectangle(blank.copy(), (30,30), (370,370), 255, -1)
weird_shape = cv.bitwise_and(circle,rectangle)
cv2_imshow(weird_shape)
```

```python
weird_masked_cats = cv.bitwise_and(cats, cats, mask=weird_shape)
cv2_imshow(weird_masked_cats)
```
