```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
from google.colab.patches import cv2_imshow
```

```python
cats = cv.imread("/content/drive/MyDrive/Colab Notebooks/opencv_for_python/Resources/Photos/cats.jpg")
cv2_imshow(cats)
cv.waitKey(0) 
cv.destroyAllWindows()
```

```python
gray = cv.cvtColor(cats, cv.COLOR_BGR2GRAY)
cv2_imshow(gray)
```

```python
# Grayscale Histogram
gray_hist = cv.calcHist([gray], [0], None, [256], [0,256])
plt.figure()
plt.title('Grayscale Histogram')
plt.xlabel('Bins')
plt.ylabel('# of pixels')
plt.plot(gray_hist)
plt.xlim([0,256])
plt.show()
```

```python
# Color Histogram
```

```python
cats = cv.imread("/content/drive/MyDrive/Colab Notebooks/opencv_for_python/Resources/Photos/cats.jpg")
cv2_imshow(cats)
cv.waitKey(0) 
cv.destroyAllWindows()
```

```python
blank = np.zeros(cats.shape[:2], dtype = 'uint8')
circle = cv.circle(blank, (cats.shape[1]//2, cats.shape[0]//2), 100, 255, -1)
mask = cv.bitwise_and(cats, cats, mask=circle)
cv2_imshow(mask)
```

```python
plt.figure()
plt.title('Color Histogram')
plt.xlabel('Bins')
plt.ylabel('# of pixels')
colors = ('b','g','r')
for i, col in enumerate(colors):
  hist = cv.calcHist([cats],[i], None, [256], [0,256])
  plt.plot(hist, color = col)
  plt.xlim([0,256])
plt.show()
```

```python
blank = np.zeros(cats.shape[:2], dtype = 'uint8')
mask = cv.circle(blank, (cats.shape[1]//2, cats.shape[0]//2), 100, 255, -1)
masked = cv.bitwise_and(cats, cats, mask=mask)
cv2_imshow(masked)
```
