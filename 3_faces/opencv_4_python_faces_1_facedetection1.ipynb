```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
from google.colab.patches import cv2_imshow
```

```python
lady = cv.imread("/content/drive/MyDrive/Colab Notebooks/opencv_for_python/Resources/Photos/lady.jpg")
cv2_imshow(lady)
cv.waitKey(0) 
cv.destroyAllWindows()
```

```python
gray = cv.cvtColor(lady, cv.COLOR_RGB2GRAY)
cv2_imshow(gray)
```

```python
haar_cascade = cv.CascadeClassifier("/content/drive/MyDrive/Colab Notebooks/opencv_for_python/3_faces/haarcascade_frontalface_default.xml")
```

```python
faces_rect = haar_cascade.detectMultiScale(gray, scaleFactor = 1.1, minNeighbors = 3)
```

```python
print(f"Number of faces found = {len(faces_rect)}")
```

```python
for (x,y,w,h) in faces_rect:
  cv.rectangle(lady, (x,y),(x+w,y+h), (0,255,0), thickness=2)
cv2_imshow(lady)
```

```python
group = cv.imread("/content/drive/MyDrive/Colab Notebooks/opencv_for_python/Resources/Photos/group 1.jpg")
cv2_imshow(group)
cv.waitKey(0) 
cv.destroyAllWindows()
```

```python
gray = cv.cvtColor(group, cv.COLOR_RGB2GRAY)
cv2_imshow(gray)
```

```python
faces_rect = haar_cascade.detectMultiScale(gray, scaleFactor = 1.1, minNeighbors = 1)
```

```python
print(f"Number of faces found = {len(faces_rect)}")
```

```python
for (x,y,w,h) in faces_rect:
  cv.rectangle(group, (x,y),(x+w,y+h), (0,255,0), thickness=2)
cv2_imshow(group)
```
